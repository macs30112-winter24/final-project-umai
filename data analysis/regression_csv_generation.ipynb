```python
import pandas as pd

#Chicago part
chi_data_output = pd.read_csv('C:/Users/13945/Desktop/data mgmt/umai/CHI_DATA_OUTPUT.csv')
all_demo = pd.read_csv('C:/Users/13945/Desktop/data mgmt/umai/all_demo.csv')
all_poverty = pd.read_csv('C:/Users/13945/Desktop/data mgmt/umai/all_poverty.csv')

chi_data_output['house.age'] = 2023 - chi_data_output['YR_BUILT']

def clean_location(location):
    if 'CHI - ' in location:
        return location.split('CHI - ')[1]
    else:
        return location

# Apply the corrected function to the 'LOCATION' column to create the 'cleaned_neighborhood' variable
chi_data_output['neighborhood'] = chi_data_output['LOCATION'].apply(clean_location)

# Check the unique values to confirm the changes
chi_data_output['neighborhood'].unique()

# Filter out the all_poverty and all_demo datasets for 'Chicago' only
all_poverty_chicago = all_poverty[all_poverty['city'] == 'Chicago']
all_demo_chicago = all_demo[all_demo['city'] == 'Chicago']

# Merge the datasets based on 'neighborhood' and 'community'
chi_data_output_merged = chi_data_output.merge(all_poverty_chicago, left_on='neighborhood', right_on='community', how='inner')
chi_data_output_final = chi_data_output_merged.merge(all_demo_chicago, on='community', how='inner', suffixes=('_poverty', '_demo'))

#Add city name
chi_data_output_final['city']= 'Chicago'

# Selecting only the required columns for regression analysis
columns_for_regression = [
    'PRICE', 'SQFT', 'house.age', 'PROP_TYPE', 'PARK_DIST', 'HOSPITAL_D', 'RAIL_DIST', 
    'BUS_COUNTS', 'FnB_COUNTS', 'neighborhood', 'population.density', 'median.age', 'Dominant Race','city'
]
chi_data_final_for_regression = chi_data_output_final[columns_for_regression]

# Checking the final dataframe before regression analysis
chi_data_final_for_regression.head()
```

```python
# Save the Chicago final dataset to CSV
chi_data_final_for_regression.to_csv('C:/Users/13945/Desktop/data mgmt/umai/chi_final_for_regression.csv', index=False)
```

```python
#LA part
la_data_output = pd.read_csv('C:/Users/13945/Desktop/data mgmt/umai/LAX_DATA_OUTPUT.csv')
all_demo = pd.read_csv('C:/Users/13945/Desktop/data mgmt/umai/all_demo.csv')
la_data_output['house.age'] = 2023 - la_data_output['YR_BUILT']

#Merge the datasets
la_merged_demo = la_data_output.merge(all_demo, left_on='neighborhood', right_on='community', how='inner')
la_merged_demo['city'] = 'Los Angeles City'

#Preview LA dataset
la_final_dataset = la_merged_demo[columns_for_regression]
la_final_dataset.head()
```

```python
# Combine city csv
chi_final = pd.read_csv('C:/Users/13945/Desktop/data mgmt/umai/chi_final_for_regression.csv')
lax_final = pd.read_csv('C:/Users/13945/Desktop/data mgmt/umai/lax_final_for_regression.csv')
nyc_final = pd.read_csv('C:/Users/13945/Desktop/data mgmt/umai/nyc_final_for_regression.csv')

# Concatenate the datasets
combined_final = pd.concat([chi_final, lax_final, nyc_final], ignore_index=True)

# Checking the final dataframe before regression analysis
combined_final.head()
```

```python
# Save the combined dataset to a CSV file
combined_final.to_csv('C:/Users/13945/Desktop/data mgmt/umai/combined_final_for_regression.csv', index=False)
```

```python
#Neighborhood level csv generation
combined_final = pd.read_csv('C:/Users/13945/Desktop/data mgmt/umai/combined_final_for_regression.csv')

# Calculate mean for specified variables
mean_columns = ['SQFT', 'house.age', 'PARK_DIST', 'HOSPITAL_D', 'RAIL_DIST', 'BUS_COUNTS', 'FnB_COUNTS', 'population.density', 'median.age']
means = combined_final.groupby(['city', 'neighborhood'])[mean_columns].mean().reset_index()

# Calculate mode for categorical variables
modes = combined_final.groupby(['city', 'neighborhood']).agg(lambda x: pd.Series.mode(x)[0] if not x.mode().empty else None)[['PROP_TYPE', 'Dominant Race']].reset_index()

# Merge means and modes
neighborhood_level_regression = pd.merge(means, modes, on=['city', 'neighborhood'])

# Calculate mean PRICE
neighborhood_level_regression['average_PRICE'] = combined_final.groupby(['city', 'neighborhood'])['PRICE'].mean().reset_index(drop=True)

#Preview neighborhood level dataset
neighborhood_level_regression.head()
```

```python
# Save the neighborhood-level dataset
neighborhood_level_regression.to_csv('C:/Users/13945/Desktop/data mgmt/umai/neighborhood_level_regression.csv', index=False)
```
