```python
combined_final = pd.read_csv('C:/Users/13945/Desktop/data mgmt/umai/combined_final_for_regression.csv')

# Calculate mean for specified variables
mean_columns = ['SQFT', 'house.age', 'PARK_DIST', 'HOSPITAL_D', 'RAIL_DIST', 'BUS_COUNTS', 'FnB_COUNTS', 'population.density', 'median.age']
means = combined_final.groupby(['city', 'neighborhood'])[mean_columns].mean().reset_index()

# Calculate mode for categorical variables
modes = combined_final.groupby(['city', 'neighborhood']).agg(lambda x: pd.Series.mode(x)[0] if not x.mode().empty else None)[['PROP_TYPE', 'dominant.race']].reset_index()

# Merge means and modes
neighborhood_level_regression = pd.merge(means, modes, on=['city', 'neighborhood'])

# Calculate mean PRICE
neighborhood_level_regression['average_PRICE'] = combined_final.groupby(['city', 'neighborhood'])['PRICE'].mean().reset_index(drop=True)

# Save the neighborhood-level dataset
neighborhood_level_regression.to_csv('C:/Users/13945/Desktop/data mgmt/umai/neighborhood_level_regression.csv', index=False)
```
